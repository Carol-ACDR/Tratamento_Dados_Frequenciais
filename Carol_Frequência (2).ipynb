Primeiramente importar o pandas e o Numpy, depois ler o csv, de forma que todos os valores sejam lidos como Valores(header= None), usando a separação por ';' e tornando todas as colunas de volta ao tipo float utilizando do decimal =','. Para tornar todas as colunas em uma só, foi utilizado o método 'melt'. E assim, ocorrerem duas contagem de valores utilizando o value_counts. Uma contagem para o valor Absoluto e outra para o Relativo. Após o concatenamento dessas a um único DataFrame. Foi feito um último Dataframe destinado a ser o valor total, juntando a soma dos valores Absolutos de um lado e a soma dos Valores relativos de outro. Assim, esse último dataframe foi adicionado como row final, mostrando a soma dos valores totais.

```python
import pandas as pd
import numpy as np
D1 = pd.read_csv('dados1.csv',sep =';',header=None,decimal =',')
D1
```

```python
F1 = pd.melt(D1,value_vars=None)
X = F1['value'].value_counts()
X1 = X.sort_index().rename('Frequencia Absoluta',axis =1)
Z = F1['value'].value_counts(normalize = True)*100
Z1 = Z.sort_index().rename('Frequencia Relativa',axis =1)
T1 = pd.concat([X1,Z1],axis = 1)
T2 = T1.append(pd.DataFrame([[X1.sum(),Z1.sum()]], index =['Total'],columns = ('Frequencia Absoluta','Frequencia Relativa'),dtype = int))
T2
```
